```python
# A program to generate a triangle of the user's desire:

row = int(input("Please enter the row size of the triangle: "))

symbol = input("Please enter a symbol to draw the triangle: ")

#Right angle triangle:
for a in range (1,row+1):
    for b in range (1,a+1):
        print(symbol,end="")
    print("")
print("")

#Right angle triangle where the space is implimented on second loop
for aa in range (1,row+1):
    for bb in range (1,aa+1):
        print(symbol,end=" ")
    print("")
print("")

#right angle triangle from the right:

for c in range (1,row+1):
    for d in range (row-c):
        print(" ", end="")
    for e in range (1,c+1):
        print(symbol,end="")
    print("")
print("")

#right angle triangle with spaces from the right:

for cc in range (1,row+1):
    for dd in range (row-cc):
        print(" ", end=" ")
    for ee in range (1,cc+1):
        print(symbol,end=" ")
    print("")
print("")


#Icoseles triangle:

for f in range (1,row+1):
    for g in range (row-f):
        print(" ", end="")
    for h in range (1,2*f):
        print(symbol, end="")
    print("")
print("")

#Icoseles triangle with spaces:

for ff in range (1,row+1):
    for gg in range (row-ff):
        print(" ", end="")
    for hh in range (1,ff+1):
        print(symbol, end=" ")
    print("")
print("")

#upside down right angle triangle to the left:

for i in range (1, row+1):
    for j in range(0,i-1):
        print("", end="")
    for k in range (row+1-i):
        print(symbol,end="") 
    print("")
print("")

#upside down right angle triangle to the left with spaces:

for ii in range (1, row+1):
    for jj in range(0,ii-1):
        print("", end="")
    for kk in range (row+1-ii):
        print(symbol,end=" ") 
    print("")
print("")

#upside down right angle triange to the right:

for l in range (1, row+1):
    for m in range(0,l-1):
        print(" ", end="")
    for n in range (row+1-l):
        print(symbol,end="")  
    print("")
print("")

#upside down right angle triangle to the right with spaces:

for ll in range (1, row+1):
    for mm in range(0,ll-1):
        print("  ", end="")
    for nn in range (row+1-ll):
        print(symbol,end=" ") 
    print("")
print("")

#upside down triange:

for p in range (0, row):
    for q in range(1,p+1):
        print(" ", end="")
    for r in range (2*(row-p)-1):
        print(symbol,end="")  
    print("")
print("")

#upside down triange with spaces:
for pp in range (1, row+1):
    for qq in range(0,pp-1):
        print(" ", end="")
    for rr in range (row+1-pp):
        print(symbol,end=" ")  
    print("")
print("")

#Upside down triangle where first loop is iterating rows, second loop is iterating spaces for column before symbol printing and third loop prints symbols
#Indent then triangle due to y loop going from 0, x iteration:

for x in range (1, row+1):
    for y in range(0,x):
        print(" ", end="")
    for z in range (row+1-x):
        print(symbol,end=" ") 
    print("")
print("")

```
